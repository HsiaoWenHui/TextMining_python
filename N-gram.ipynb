```python
from openpyxl import load_workbook
import codecs
#處理編碼的套件
import operator
##處理字典檔排序的套件
import re

wb = load_workbook('hw1_text.xlsx')
sheet = wb.active 
cutlist = "「」<>/:：;；,、＂’，.。！？｢\"\'\\\n\r《》“”!@#$%^&*()0123456789（）abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"

termList=[]
```

```python
term1='銀行'
term2='信用卡'
term3='匯率'
term4='台積電'
term5='台灣'
term6='日本'
term1List = []
term2List = []
term3List = []
term4List = []
term5List = []
term6List = []
```

```python
def deDot(line):
        sentence = ""
        for word in line:
            if word not in cutlist: #如果文字不是標點符號，就把字加到句子中
                sentence += word
        return(sentence)
```

```python
for row in range(2,sheet.max_row-1):
    for column in"E":
        cell_name = "{}{}".format(column, row)
        text=sheet[cell_name].value
        if(re.match(term1, text)):
            text=deDot(text)
            term1List.append(text)
        elif(re.match(term2, text)):
            text=deDot(text)
            term2List.append(text)
        elif(re.match(term3, text)):
            text=deDot(text)
            term3List.append(text)
        elif(re.match(term4, text)):
            text=deDot(text)
            term4List.append(text)
        elif(re.match(term5, text)):
            text=deDot(text)
            term5List.append(text)
        elif(re.match(term6, text)):
            text=deDot(text)
            term6List.append(text)
print("the total : ",sheet.max_row)
print("term1 : ",len(term1List))
print("term2 : ",len(term2List))
print("term3 : ",len(term3List))
print("term4 : ",len(term4List))
print("term5 : ",len(term5List))
print("term6 : ",len(term6List))
```

```python
def ngram(textLists,n,minFreq):
    words=[]
    words_freq={}
    result=[]
    for textList in textLists:
        for w in range(len(textList)-(n-1)):
            words.append(textList[w:w+n])
    
    for word in words:
        if word not in words_freq:
            words_freq[word]=words.count(word)
    words_freq=sorted(words_freq.items(),key=operator.itemgetter(1),reverse=True)
    for word in words_freq:
        if word[1] >= minFreq:
            result.append(word)
    return result
```

```python
def show(textList):
    i=0
    while i<100:
        print(textList[i][0],textList[i][1])
        i+=1
```

```python
bank=ngram(term1List,2,2)
print("銀行主題的前100名Keywords:")
show(bank)
```

```python
isms=ngram(term4List,3,2)
print("台積電主題的前100名Keywords:")
show(isms)
```

```python
taiwan=ngram(term5List,2,2)
print("台灣主題的前100名Keywords:")
show(taiwan)
```

```python
japan=ngram(term6List,2,2)
print("日本主題的前100名Keywords:")
show(japan)
```
