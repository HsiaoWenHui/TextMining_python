```python
import math
import pandas as pd
from operator import itemgetter
```

```python
raw_terms=pd.read_csv('term.csv')
raw_data=pd.read_csv('text.csv')
```

```python
terms=list(raw_terms['term'])
data=list(raw_data['標題']+raw_data['內容'])
print("共有 {} 個字".format(len(terms)))
print("共有 {} 筆data".format(len(data)))

del raw_terms
```

```python
word2id={}
id2word=[]
#i是index 使用enumerate(), 它會傳回一個次序及值成組tuple的疊代
for i,term in enumerate(terms):
    word2id[term]=i
    id2word.append(term)
del terms
```

```python
#算每個詞出現在幾篇文章中
df=[]
for term in id2word:
    df.append(0)
    for d in data:
        if term in d:
            df[word2id[term]]+=1
```

```python
def MI(f_x,f_y,f_xy,N):
    return math.log(f_xy/(f_x*f_y))
```

```python
def lift(f_x,f_y,f_xy,N):
    return(f_xy/f_x)/(f_y/N)
```

```python
def count_co_occur(data,term_x,term_y):
    cnt=0
    for d in data:
        if term_x in d and term_y in d :
            cnt+=1
    return cnt
def count_N(data,term_x,term_y):
    cnt=0
    for d in data:
        if term_x in d or term_y in d :
            cnt+=1
    return cnt
```

```python
target_word='鴻海'
data_lift=[]
data_MI=[]
for term in id2word:
    f_x=df[word2id[target_word]]
    f_y=df[word2id[term]]
    f_xy=count_co_occur(data,target_word,term)
    N=count_N(data,target_word,term)
    data_lift.append((term,lift(f_x,f_y,f_xy,N)))
    data_MI.append((term,MI(f_x,f_y,f_xy,N)))
```
